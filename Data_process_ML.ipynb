```python
import pandas as pd 
import numpy as np
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error
from sklearn.model_selection import train_test_split
from sklearn.model_selection import cross_validate
```

# Data import and analysis 

```python
base = pd.read_csv('./base.csv')
svf1 = pd.read_csv('./svf1.csv')
svf2 = pd.read_csv('./svf2.csv')
sl1 = pd.read_csv('./sl1.csv')
sl2 = pd.read_csv('./sl2.csv')
```

# Data manipulation

### Extraction and averaging of Shadowmap

```python
sl1_out = sl1.iloc[:, 3::4]
x = sl1_out[sl1_out > 1]
x.mean()
mean_sl1 = pd.DataFrame([x.mean()]).round(2)
mean_sl1
```

```python
sl2_out = sl2.iloc[:, 3::4]
x = sl2_out[sl2_out > 1]
x.mean()
mean_sl2 = pd.DataFrame([x.mean()]).round(2)
mean_sl2
```

```python
mean_sl_total = pd.concat([mean_sl1, mean_sl2], axis = 1)
mean_sl_total
```

### Extraction and averaging of SVF

```python
svf1_out = svf1.iloc[:, 3::4]
x = svf1_out[svf1_out > 1]
x.mean()
mean_svf1 = pd.DataFrame([x.mean()]).round(2)
mean_svf1
```

```python
svf2_out = svf2.iloc[:, 3::4]
x = svf2_out[svf2_out > 1]
x.mean()
mean_svf2 = pd.DataFrame([x.mean()]).round(2)
mean_svf2
```

```python
mean_svf_total = pd.concat([mean_svf1, mean_svf2], axis = 1)
mean_svf_total
```

### Rotation and joining with Base table

```python
mean_slx= mean_sl_total.T
mean_sl3 = mean_slx.rename(columns={0: "SL %"})

mean_svfx = mean_svf_total.T
mean_svf3 = mean_svfx.rename(columns={0: "SVF %"})

mean_svf4 = mean_svf3.reset_index(drop=True)
mean_sl4 = mean_sl3.reset_index(drop=True)
result = pd.concat([base, mean_sl4, mean_svf4],axis=1)
result = result.drop_duplicates()
result.to_csv('result.csv')
result
```

## Renaming

```python
result.columns = range(result.shape[1])
clean_df = result.rename(columns={31: "sl", 32: "svf"})
clean_df
```

# Machine learning

### Predict sunlist

```python
result1 = clean_df.dropna()
#shuffle and reset index
clean_df1 =result1.sample(frac=1)
clean_df2 = clean_df1.reset_index(drop=True)

train = clean_df2[0:950]
test = clean_df2[-50:]

y_train = train['sl']
#print(y_train.describe())
X_train = train.iloc[:,:22]
y_test = test['sl']
X_test = test.iloc[:,:22]

regr = RandomForestRegressor()
regr.fit(X_train,y_train)
prediction = regr.predict(X_test)

y_test = y_test.reset_index(drop=True)
print("Mean absolute error for Sunlit prediction:",mean_absolute_error(y_test, prediction))
res1 = pd.concat([pd.DataFrame(prediction, columns=['Predicted sunlit']),y_test],axis=1)
```
