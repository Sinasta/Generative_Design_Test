# Data manipulation

### Extraction and averaging of Sunlit

```python
sl_out = sl.iloc[:, 3::4]
x = sl_out[sl_out > 1]
x.mean()
mean_sunlit = pd.DataFrame([x.mean()]).round(2)
mean_sunlit
```

### Extraction and averaging of SVF

```python
svf_out = svf.iloc[:, 3::4]
x = svf_out[svf_out > 1]
x.mean()
mean_svf = pd.DataFrame([x.mean()]).round(2)
mean_svf
```

### Rotation and joining with Base table

```python
mean_sunlitx= mean_sunlit.T
mean_sunlit1 = mean_sunlitx.rename(columns={0: "Sunlit %"})

mean_svfx = mean_svf.T
mean_svf1 = mean_svfx.rename(columns={0: "SVF %"})

mean_svf2 = mean_svf1.reset_index(drop=True)
mean_sunlit2 = mean_sunlit1.reset_index(drop=True)
result = pd.concat([base,mean_sunlit2, mean_svf2],axis=1)
#result.to_csv('result.csv')
result
```

```python
result1 = clean_df.dropna()
#shuffle and reset index
clean_df1 =result1.sample(frac=1)
clean_df2 = clean_df1.reset_index(drop=True)

train = clean_df2[0:950]
test = clean_df2[-50:]

y_train = train['sunlit']
#print(y_train.describe())
X_train = train.iloc[:,:22]
y_test = test['sunlit']
X_test = test.iloc[:,:22]

regr = RandomForestRegressor()
regr.fit(X_train,y_train)
prediction = regr.predict(X_test)

y_test = y_test.reset_index(drop=True)
print("Mean absolute error for Sunlit prediction:",mean_absolute_error(y_test, prediction))
res1 = pd.concat([pd.DataFrame(prediction, columns=['Predicted sunlit']),y_test],axis=1)
```

```python
result1 = clean_df.dropna()
#shuffle and reset index
clean_df1 = result1.sample(frac=1)
clean_df2 = clean_df1.reset_index(drop=True)

train = clean_df2[0:950]
test = clean_df2[-50:]

y_train = train['svf']
X_train = train.iloc[:,:23]
y_test = test['svf']
X_test = test.iloc[:,:23]

regr = RandomForestRegressor()
regr.fit(X_train,y_train)
prediction = regr.predict(X_test)

y_test = y_test.reset_index(drop=True)
print("Mean absolute error for SVF prediction:",mean_absolute_error(y_test, prediction))
res2 = pd.concat([pd.DataFrame(prediction, columns=['Predicted svf']),y_test],axis=1)
```
